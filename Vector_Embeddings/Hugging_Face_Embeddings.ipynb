```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    show_progress = True
)
```

```python
embeddings
```

```python
Text = "Hi my name is venkata Dharaneswara Reddy, This is Day 2 of Learning RAG"

embedding = embeddings.embed_query(Text)
print(f"text : {Text}")
print(f"Embedding Length : {len(embedding)}")
print(embedding)
```

```python
sentences = [
    "The cat sat on the mat",
    "The cat sat on the mat",
    "The dog played in the yard",
    "I love programming in Python",
    "Python is my favorite programming language"
]

embedding_sentence=embeddings.embed_documents(sentences)

print(embedding_sentence[0])
print(embedding_sentence[1])
print(embedding_sentence[2])
```
